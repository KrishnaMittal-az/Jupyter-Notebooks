```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:\PycharmProjects\ML Algo\Jup notebooks\CSV\Salary.csv')
print(data)
```

```python
print(data.isnull().sum())
```

```python
X = data["YearsExperience"]
Y = data["Salary"]
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
m = 0
c = 35000
L = 0.0001
n = len(X)
epochs = 10000
```

```python
for i in range(epochs):
    Y_pred = m*X+c
    D_m = (-2/n) * sum(X * (Y - Y_pred))
    D_c = (-2/n) * sum(Y - Y_pred)
    m = m - L * D_m
    c = c + L * D_c
print(m , c)
```

```python
# Make Predictions
Y_pred = m*X+c
print(Y_pred)
```

```python
plt.scatter(X,Y,color = "red" , label = "Data Points")
plt.plot(X,Y_pred , color = "blue" , label = "Predicted Salary")
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.legend()
plt.title("Linear Regression")
plt.show()
```

```python
squared_error = (Y - Y_pred)**2
mse = squared_error.mean()
```
